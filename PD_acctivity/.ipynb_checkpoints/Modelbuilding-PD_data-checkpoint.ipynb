```python
%matplotlib inline
from sklearn.linear_model import LogisticRegression
from sklearn import svm, metrics
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix
import itertools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
import datetime
import progressbar
from time import sleep
import scipy.signal as signal 
from detect_peaks import detect_peaks
from collections import Counter
sns.set(style="darkgrid")
from Build_LSTM import *
```

```python

def get_class_weights(y):
    counter = Counter(y)
    majority = max(counter.values())
    return  {cls: float(majority/count) for cls, count in counter.items()}
```

```python

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=60)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
df = pd.read_csv("modelbuilding/data/preprocessed/data.csv", index_col=0)
df.dropna(axis=0, how='any', inplace=True)
```

```python
df['label'] = df['label'].apply(f)
```

```python
def f(x):
    if x in [1,2,3]:
        return 1
    else:
        return 0
```

```python
sns.countplot(x="label",data=df)
plt.title("Label Distribution")
plt.xlabel("Count")
plt.ylabel("Label")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python

X_training = df[df.columns[:-2]].values
y_training = df['label'].values
    
# Calculate class weight
class_weight = get_class_weights(y_training)
    
scaler = StandardScaler(with_mean=True,with_std=True)
```

```python
X_train_scaled = scaler.fit_transform(X=X_training)
```

```python
model = LogisticRegression(class_weight=class_weight, penalty='l2')
model.fit(X_train_scaled, y_training)
```

```python
#LOO-Validation on each subject
subject_list = np.unique(df['subject_id'])
y_test_all = []
y_pred_all = []
iterations = list(itertools.combinations(subject_list, 1))
size = len(iterations) # to speed up simulations
it_index = np.random.choice(len(iterations), size=size, replace = False)
# initialization of Bar 
bar = progressbar.ProgressBar(maxval=size, \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
bar.start()
bar.update(0)
bar_i = 0
for i in it_index: # two combination of tested subjects
    
    subject = iterations[i]
    #print("==================")
    #print('Model tested on subjects with no : ', subject)
    df_training = df[np.logical_and(df['subject_id']!=subject[0],True)]#,df['subject_id']!=subject[1] )]
    df_test = df[np.logical_or(df['subject_id']==subject[0],False )]#df['subject_id']==subject[1] )]
    #print("Test set and training set proportion : ",float(df_test.shape[0])/df_training.shape[0] )
    X_test = df_test[df_test.columns[:-2]].values
    y_test = df_test['label'].values
    X_training = df_training[df_training.columns[:-2]].values
    y_training = df_training['label'].values
    
    # Calculate class weight
    class_weight = get_class_weights(y_training)
    
    
    # model definition
    model = LogisticRegression(class_weight=class_weight, penalty='l2')
#     model = svm.SVC(gamma=0.001, class_weight = class_weight)
    model.fit(X_training, y_training)
    y_pred = model.predict(X_test)
    
    y_test_all.extend(list(y_test))
    y_pred_all.extend(list(y_pred))
    #print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
    bar.update(bar_i+1)
    bar_i += 1
bar.finish()
```

```python
label_dict = {1:'walking',
             2:'walking_upstairs',
             3:'walking_downstairs',
             4:'sitting',
             5:'standing',
             6:'laying'}
class_names = list(label_dict.values())
```

```python
# save_with_name = "Results/EQ_SIZED_64_logreg-with-weights" ################################### 
cnf_matrix = confusion_matrix(y_test_all, y_pred_all)
np.set_printoptions(precision=2)
# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')
# plt.savefig(save_with_name+'.png')
plt.show()
```

```python
true_pos = np.diag(cnf_matrix).sum()
accuracy = true_pos/float(cnf_matrix.sum())
print('Overall accuracy of the Model', accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_all, y_pred_all,target_names=class_names))
```

```python
def classifaction_report_csv(report, filename):
    class MagicString(str):
        magicSplit = str.split
    report_data = []
    lines = report.split('\n')
    for line in lines[2:-3]:
        row = {}
        s = MagicString(line)
        row_data = s.magicSplit()
        print(row_data)
        row['class'] = row_data[0]
        row['precision'] = float(row_data[1])
        row['recall'] = float(row_data[2])
        row['f1_score'] = float(row_data[3])
        row['support'] = float(row_data[4])
        report_data.append(row)
    dataframe = pd.DataFrame.from_dict(report_data)
    dataframe.to_csv(filename, index = False)

```
